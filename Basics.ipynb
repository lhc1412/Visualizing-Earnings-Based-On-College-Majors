# Visualizing Earnings Based On College Majors

In this project, We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors). 

Using visualizations, we can start to explore questions from the dataset like:
- Do students in more popular majors make more money?
- How many majors are predominantly male? Predominantly female?
- Which category of majors have the most students?

### Summary of results
After analysing the data, the following findings can be drawn.

- Students in more popular majors do not make more money.
- Students that majored in subjects that were majority female do not make more money.
- No link between the number of full-time employees and median salary is found.
- 55 subjects are predominantly male.
- 69 subjects are predominantly female.
- In top 10 majors with highest median earnings, 8 majors are predominantly male.
- In last 10 majors with highest median earnings, all 10 majors are predominantly female.
- No significant different in unemployment rate can be found between top 10 and last 10 majors.
- Arts, Health, and Education are the major categories with highest number of students.

## Reading in data

```python
# read in data
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
%matplotlib inline

# explore the data primarily
recent_grads = pd.read_csv("recent-grads.csv")
print("First Row:\n",recent_grads.iloc[0],"\n")
print()
print("Head 5:\n",recent_grads.head(),"\n")
print()
print("Tail 5:\n",recent_grads.tail(),"\n")
print()
print("Description:\n",recent_grads.describe(),"\n")
```

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Descriptions of column headers are following:

- Rank:	Rank by median earnings
- Major_code:	Major code, FO1DP in ACS PUMS
- Major	Major: description
- Major_category:	Category of major from Carnevale et al
- Total:	Total number of people with major
- Sample_size:	Sample size (unweighted) of full-time, year-round ONLY (used for earnings)
- Men:	Male graduates
- Women:	Female graduates
- ShareWomen:	Women as share of total
- Employed:	Number employed (ESR == 1 or 2)
- Full_time:	Employed 35 hours or more
- Part_time:	Employed less than 35 hours
- Full_time_year_round:	Employed at least 50 weeks (WKW == 1) and at least 35 hours (WKHP >= 35)
- Unemployed:	Number unemployed (ESR == 3)
- Unemployment_rate:	Unemployed / (Unemployed + Employed)
- Median:	Median earnings of full-time, year-round workers
- P25th:	25th percentile of earnings
- P75th:	75th percentile of earnings
- College_jobs:	Number with job requiring a college degree
- Non_college_jobs:	Number with job not requiring a college degree
- Low_wage_jobs:	Number in low-wage service jobs

## Data Analysis

### Relationship between number of student and salary

In this part, we would like to explore following questions.
- Do students in more popular majors make more money?
- Do students that majored in subjects that were majority female make more money?
- Is there any link between the number of full-time employees and median salary?

#### Fig 1 Sample_size vs. Median

```python
# plot scatter diagram Sample_size vs. Median
recent_grads.plot(x='Sample_size', y='Median', 
                  kind='scatter', title='Sample_size vs. Median', figsize=(10,5))
```

Fig 1 shows no correlation between Number of full-time student and Median salary of full-time, year-round workers.

```python
# plot scatter diagram Sample_size vs. Unemployment_rate
recent_grads.plot(x='Sample_size', y='Unemployment_rate',
                  kind='scatter', title='Sample_size vs. Unemployment_rate', figsize=(10,5))
```

Fig 2 shows no correlation between Number of full-time student and Unemployment rate.

#### Fig 3 Full_time vs. Median

```python
# plot scatter diagram Full_time vs. Median
recent_grads.plot(x='Full_time', y='Median',
                  kind='scatter', title='Full_time vs. Median', figsize=(10,5))
```

Fig 3 shows no correlation between Number of full-time student and Median salary of full-time, year-round workers.

#### Fig 4 ShareWomen vs. Unemployment_rate

```python
# plot scatter diagram ShareWomen vs. Unemployment_rate
recent_grads.plot(x='ShareWomen', y='Unemployment_rate',
                  kind='scatter', title='ShareWomen vs. Unemployment_rate', figsize=(10,5))
```

Fig 4 shows no correlation between Women as share of total and Unemployment rate.

#### Fig 5 Men vs. Median

```python
# plot scatter diagram Men vs. Median
recent_grads.plot(x='Men', y='Median',
                  kind='scatter', title='Men vs. Median', figsize=(10,5))
```

Fig 5 shows no correlation between Number of Male graduates and the Median salary of full-time, year-round workers.

#### Fig 6 Women vs. Median

```python
# plot scatter diagram Women vs. Median
recent_grads.plot(x='Women', y='Median',
                  kind='scatter', title='Women vs. Median', figsize=(10,5))
```

Fig 6 shows no correlation between Number of Female graduates and the Median salary of full-time, year-round workers.

#### Fig 7 Total vs. Median

```python
# plot scatter diagram Total vs. Median
recent_grads.plot(x='Total', y='Median',
                  kind='scatter', title='Total vs. Median', figsize=(10,5))
```

#### Fig 8 scatter_matrix of Sample_size vs. Median

```python
# plot scatter_matrix of Sample_size vs. Median
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,10))
```

#### Fig 9 scatter_matrix of Sample_size, Median and Unemployment rate

From the above figures, we can see no relationship between the popularity of a major and the salary. The following findings are drawn: 
- Students in more popular majors do not make more money.
- Students that majored in subjects that were majority female do not make more money.
- No link between the number of full-time employees and median salary is found.

### Distribution of Students

#### Distribution of different data

```python
# create fig and ax for plots
fig, ax = plt.subplots(nrows=4, ncols=2, figsize=(20,30))
cols = ['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']

# create plots
for i in range(0,8):
    ax = fig.add_subplot(4,2,i+1)
    ax = recent_grads[cols[i]].hist(bins=30, range=(0,max(recent_grads[cols[i]])))
    ax.set_xlabel(cols[i])
```

From the Histogram of ShareWomen, 55 subjects are with less than 40% female. These subjects are predominantly male. 

69 subjects are with more than 60% female. These subjects are predominantly female.
